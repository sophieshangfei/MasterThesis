```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
%matplotlib inline
```

```python
#import csv
#pd.set_option('display.max_columns', 30)
#pd.set_option('display.max_rows', None)
df = pd.read_csv('akshay.csv', sep = ",")
df.head(400)
```

```python
#get rid of scientific notation
pd.set_option('display.float_format', lambda x: '%.3f' % x)

#replace all null and nan to the np.nan
df = df.replace('null', np.nan)
df.replace('nan', np.nan)

#turn object into floats and round them
df['end_time'] = round(df['end_time'].astype(float))
df['start_time'] = round(df['start_time'].astype(float))
df['builtin_rt'] = df['builtin_rt'].astype(float) #dont round up builtin reaction time
df['rod fixing checked'] = df['rod fixing checked'].astype(float)
```

```python
df.head() #only look at the first 10 rows of the data frame
```

```python
#select columns
data = df[['price_displayed', 'key pressed','environment', 'order', 
           'round_instance', 'task', 'function id', 'rod fixing checked', 
           'builtin_rt', 'end_time', 'start_time']]
```

```python
data.head()
```

```python
def latency(lst):
    i = 0
    latency_s = []
    while i < len(lst)-1:
        t = float(lst[i+1]) - float(lst[i])
        latency_s.append(t)
        i = i + 1
    latency_s.append(0)
    return latency_s
```

```python
#rename columns because i am stupid
data = data.rename(index=str, columns={"round instance": "round_instance", "function id": "function_id", 
                                       "rod fixing checked": "quality_check", "builtin rt": "rt_builtin",
                                       "key pressed":"key"})
data.head()
```

```python
#get rid of all data collected from caller function
#caller functions were used to initilized elements - dont have real meaning to us
data = data[data.function_id != 'caller']
data = data[pd.notnull(data['function_id'])]
```

```python
#take a look at rod fixing data
data_rod = data[data.function_id == "rod_fixing"]

#do a data quality check
def quality_check(lst):
    check = lst.tolist()
    if 0 in check:
        return True
    else:
        return False
quality_check(data_rod['quality_check'])
```

```python
def price(lst):
    i = 0
    record = []
    while i < len(lst):
        if lst[i] not in record:
            record.append(lst[i])
            i = i + 1
        if lst[i] == lst[i - 1]:
            i = i + 1
        else:
            if record.count(lst[i]) == 1:
                record.append(lst[i])
                i = i + 1
    return record
```

```python
#just to look at fishing data
data_fish = data[(data.function_id == 'fishing_h')|(data.function_id == 'fishing_l')]

```

```python
#take a look at low environment
data_fish_l = data_fish[data_fish.function_id == "fishing_l"]
data_fish_l.head()
```

```python
#take a look at 11 and 21 seperately (11, 21 are both instances of low environment)
data_fish_11 = data_fish_l[data_fish_l.round_instance == "11"]
data_fish_21 = data_fish_l[data_fish_l.round_instance == "21"]
```

```python
#take a look at high environment
data_fish_h = data_fish[data_fish.function_id == "fishing_h"]
data_fish_12 = data_fish_h[data_fish_h.round_instance == "12"]
data_fish_22 = data_fish_h[data_fish_h.round_instance == "22"]
```

```python
# #check to see if round 1 are all the same price
price21 = price(data_fish_21["price_displayed"])
price22 = price(data_fish_22["price_displayed"])
print(price21)
print(price22)
#price(data_fish_11["price_displayed"].tolist())
```

```python
# compute latency for each environment - round pair
data_fish_12['latency'] = latency(data_fish_12["start_time"])
data_fish_11['latency'] = latency(data_fish_11["start_time"])
data_fish_22['latency'] = latency(data_fish_22["start_time"])
data_fish_21['latency'] = latency(data_fish_21["start_time"])
```

```python
#low environment 11, 21
low_price = data_fish_11['price_displayed'].append(data_fish_21['price_displayed'])
low_tap = np.arange(1, len(data_fish_11['price_displayed'])+len(data_fish_21['price_displayed'])+1)
low_latency = data_fish_11['latency'].append(data_fish_21['latency'])
```

```python
fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('tap (n)')
ax1.set_ylabel('latency (ms)')
ax1.plot(low_tap, low_latency, color=color)
ax1.set_ylim([0,3000])
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:red'
ax2.set_ylabel('price')  # we already handled the x-label with ax1
ax2.scatter(low_tap, low_price, color=color, s = 1)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

```python
#low environment 12, 22
high_price = data_fish_12['price_displayed'].append(data_fish_22['price_displayed'])
high_tap = np.arange(1, len(data_fish_12['price_displayed'])+len(data_fish_22['price_displayed'])+1)
high_latency = data_fish_12['latency'].append(data_fish_22['latency'])
```

```python
fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('tap (n)')
ax1.set_ylabel('latency (ms)')
ax1.plot(high_tap, high_latency, color=color)
ax1.set_ylim([0,3000])
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:red'
ax2.set_ylabel('price')  # we already handled the x-label with ax1
ax2.scatter(high_tap, high_price, color=color, s = 1)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
