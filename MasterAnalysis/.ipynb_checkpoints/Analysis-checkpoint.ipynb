```python
def price(lst):
    i = 0
    record = []
    while i < len(lst):
        if lst[i] not in record:
            record.append(lst[i])
            i = i + 1
        if lst[i] == lst[i - 1]:
            i = i + 1
        else:
            if record.count(lst[i]) == 1:
                record.append(lst[i])
                i = i + 1
    return record
```

```python
def latency(lst):
    i = 0
    latency_s = []
    while i < len(lst)-1:
        t = float(lst[i+1]) - float(lst[i])
        latency_s.append(t)
        i = i + 1
    latency_s.append(0)
    return latency_s
```

```python
l_11 = latency(data_rod_1_11["start_time"])
data_rod_1_11["latency"] = l_11
l_21 = latency(data_rod_1_21["start_time"])
data_rod_1_21["latency"] = l_21
l_12 = latency(data_rod_1_12["start_time"])
data_rod_1_12["latency"] = l_12
l_22 = latency(data_rod_1_22["start_time"])
data_rod_1_22["latency"] = l_22
```
