```python
def latency(ser):
    """ takes in a series and turn into a list
    then calculate latency
    the same function in pandas is called np.diff()
    """
    lst = ser.tolist()
    i = 0
    latency_s = []
    while i < len(lst)-1:
        if lst[i] != np.nan:
            t = lst[i+1] - lst[i]
            latency_s.append(t)
            i = i + 1
        else:
            latency_s.append(np.nan)
    latency_s.append(0)
    return latency_s
```

```python
def drop_first_last (ser, k):
    """
    ser is a series from your data frame
    k - how many taps to fix the rod
    return a list of lists without the first and last tap
    """
    to_ret = []
    lst = ser.tolist()
    chunks = [lst[x:x+k] for x in range(0, len(lst), k)]
    for c in chunks:
        to_ret = to_ret + c[1:-1]
    
    return to_ret
```

```python
def extract_choice(ser):
    choice = []
    for s in ser:
        c = 0
        while c < len(s):
            r = json.loads(s[c])
            for key in r.keys():
                choice.append(r[key])
            c = c + 1
    return choice
```

```python
def extract_choice(ser):
    choice = []
    for s in ser:
        c = 0
        while c < len(s):
            r = json.loads(s[c])
            for key in r.keys():
                choice.append(r[key])
            c = c + 1
    return choice
```
